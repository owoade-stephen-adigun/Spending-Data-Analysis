```python
spending_by_item = df.groupby('ITEMS')['DEBIT'].sum().sort_values(ascending=False)
spending_by_item
```

```python
spending_by_item
```

```python
spending_by_item = spending_by_item.reset_index()
spending_by_item
```

```python
spending_by_item = spending_by_item.drop([12,13,14])
spending_by_item
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(8,5))
plt.bar(spending_by_item['ITEMS'], spending_by_item['DEBIT'])
plt.title('Spending by Item')
plt.ylabel('Amount ()')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

INSIGHT
The bar chat above tells us that the bulk of the income is spent on your wife which is a responsibility to be fulfilled as the man of the house, however it should be judiciously spent over time.
Moreover, transfer, gift, grocery and foodco are items that you have control over.

```python
income_by_item = df.groupby('ITEMS')['CREDIT'].sum().sort_values(ascending=False)
income_by_item
```

```python
income_by_item = income_by_item.reset_index()
income_by_item
```

```python
plt.figure(figsize=(8,5))
plt.bar(income_by_item['ITEMS'], income_by_item['CREDIT'])
plt.title('income by Item')
plt.ylabel('Amount ()')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
